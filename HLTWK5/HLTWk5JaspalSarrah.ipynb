```python
# Import pandas library
import pandas as pd
```

```python
# Read csv file
market_data = pd.read_csv("Dow Jones Index.csv")
```

```python
# Import Altair library
import altair as alt
```

Multiple line chart showing closing price of different stock on the Dow Jones.

```python
stock_selector = alt.selection_multi(fields=[("stock")], bind="legend")

line_graph = alt.Chart(market_data).mark_line().encode(
    x = "date", 
    y = "close", 
    color = "stock:N",
    opacity=alt.condition(stock_selector, alt.value(1), alt.value(0.1)),
    tooltip = ["stock", "open", "high", "low", "close", "volume"]
    ).add_selection(stock_selector)

bar_chart = alt.Chart(market_data).mark_bar().encode(
    x="date", 
    y="volume",
    color = "stock:N"
    ).transform_filter(stock_selector)

alt.vconcat(line_graph, bar_chart)
```

Line graphs
I have chosen a line graph to show the closing prices (end of week) for different stocks. This shows how different stocks are performing over time and in comparison to reach other. In addition tool tip has been used to give key information for each stock for the corresponding week.

Stacked bar chart
I have chosen this as another key to stocks is the volume traded. The bar chart shows the volume traded each week in total and by each stock.

Selection
I have chosen a multi selector option as this will help to compare specific stocks. By selecting one or more stocks on the legend you can look at how specific stocks are performing. The two charts have been linked so that you can look at stock price and volume traded.
